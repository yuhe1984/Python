```python
try:
    f = open('D:/Jupyter/Python/Python Tutorial2/test.txt', 'r')
    print(f.read())
finally:
    if f:
        f.close()
```

```python
with open('D:/Jupyter/Python/Python Tutorial2/test.txt', 'r') as f:
    print(f.read())
```

```python
for line in f.readlines():
    print(line.strip())
```
